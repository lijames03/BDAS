```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder.appName('Data').getOrCreate()

# Read the CSV file into a DataFrame
df = spark.read.csv('Clients-data1.csv', header=True, inferSchema=True)

# Calculate summary statistics
summary = df.describe()

# Convert PySpark DataFrame to Pandas DataFrame
summary_pandas = summary.toPandas()

# Show summary statistics using Pandas
print(summary_pandas)




```

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# Create a Spark session
spark = SparkSession.builder.appName("Missing").getOrCreate()

# Read the CSV file into a DataFrame
df = spark.read.csv('Clients-data1.csv', header=True, inferSchema=True)

# Display the count of missing values for each column
missing_values_spark = df.agg(*[F.sum(F.col(c).isNull().cast("int")).alias(c) for c in df.columns])

# Convert to Pandas DataFrame and transpose
missing_values_pandas = missing_values_spark.toPandas().T

# Display the transposed result
missing_values_pandas.columns = ['Count of Missing Values']
missing_values_pandas
```
